## "Внедрение значений" - импьютация
### Обработка пропусков для количественного признака Salary

```python
# Выберем числовые колонки с пропущенными значениями
# Цикл по колонкам датасета
num_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='float64'):
        num_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
# Фильтр по колонкам с пропущенными значениями
data_num = data[num_cols]
data_num
```

```python
# Гистограмма по признакам
for col in data_num:
    plt.hist(data[col], 50)
    plt.xlabel(col)
    plt.show()
```

```python
data_num_ManagerID = data_num[['ManagerID']]
data_num_ManagerID.head()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
```

```python
# Фильтр для проверки заполнения пустых значений
indicator = MissingIndicator()
mask_missing_values_only = indicator.fit_transform(data_num_ManagerID)
mask_missing_values_only
```

```python
# Импьютация различными показателями центра распределения с помощью класса SimpleImputer 
strategies=['mean', 'median', 'most_frequent']
```

```python
def test_num_impute(strategy_param):
    imp_num = SimpleImputer(strategy=strategy_param)
    data_num_imp = imp_num.fit_transform(data_num_ManagerID)
    return data_num_imp[mask_missing_values_only]
```

```python
# Среднее значение
strategies[0], test_num_impute(strategies[0])
```

```python
# Медиана
strategies[1], test_num_impute(strategies[1])
```

```python
# Мода
strategies[2], test_num_impute(strategies[2])
```

### Обработка пропусков для категориального признака State 

```python
# Выбор категориальных колонок с пропущенными значениями
# Цикл по колонкам датасета
cat_cols = []
for col in data.columns:
    # Количество пустых значений 
    temp_null_count = data[data[col].isnull()].shape[0]
    dt = str(data[col].dtype)
    if temp_null_count>0 and (dt=='object'):
        cat_cols.append(col)
        temp_perc = round((temp_null_count / total_count) * 100.0, 2)
        print('Колонка {}. Тип данных {}. Количество пустых значений {}, {}%.'.format(col, dt, temp_null_count, temp_perc))
```

```python
# Импьютация с помощью класса SimpleImputer со стратегиями "most_frequent" или "constant".
cat_temp_data = data[['State']]
cat_temp_data.head()
```

```python
cat_temp_data['State'].unique()
```

```python
cat_temp_data[cat_temp_data['State'].isnull()].shape
```

```python
# Импьютация наиболее частыми значениями (мода)
imp2 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data_imp2 = imp2.fit_transform(cat_temp_data)
data_imp2
```

```python
# Пустые значения отсутствуют
np.unique(data_imp2)
```

```python
# Импьютация константой
imp3 = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value='N/A')
data_imp3 = imp3.fit_transform(cat_temp_data)
data_imp3
```

```python
np.unique(data_imp3)
```

```python
data_imp3[data_imp3=='N/A'].size
```

Для признака State я считаю более корректным внедрение константы N/A, так как название штата может и не соответствовать моде
